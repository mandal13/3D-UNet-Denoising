```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#####################################################################################
def data_for_plot(density_file):
    array = np.loadtxt(density_file)
    nx, ny, nz = 32, 32, 32 #get_grids(screen_file)
    
    data = np.zeros((nx, ny, nz))

    for i in range(nx):
        for j in range(ny):
            for k in range(nz):
                itmp = i + j*nz + k*nz*ny
                data[i,j,k] = array[itmp]
    return data
```

```python
path_of_dft_density = '../predict/density-target-00'
path_of_sdft_density = '../predict/density-actual-00'
path_of_predicted_density = '../predict/density-predict-00'
```

```python
data_dft = data_for_plot(path_of_dft_density)
data_sdft = data_for_plot(path_of_sdft_density)
data_pdft = data_for_plot(path_of_predicted_density)

index = 11

print(np.sum(data_dft))
print(np.sum(data_sdft))
print(np.sum(data_pdft))

print(np.sum((data_dft - data_sdft)**2)/(32**3))
print(np.sum((data_dft - data_pdft)**2)/(32**3))
print(np.sum((data_sdft - data_pdft)**2)/(32**3))
```
